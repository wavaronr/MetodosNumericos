#**Revisión Teoría de Errores**


Teniendo en cuenta lo solicitado por el presente ejercicio se define los siguientes conceptos:

En particular, el teorema establece que cualquier función suave puede aproximarse por un polinomio.(Chapra 2011)

La serie de Taylor es el metoto mediante el cual utiliza la tecnica de aproximacion a una funcion por medio de la derivada de la misma dividida en en factorial del grado de la derivada por la diferencia del valor en x menos el valor centrado para este ejercicio lo representaremos como a lo cual permite que al realizar la sumatoria de los grados de derivacion de la funcion aplicanto del teorema podemos establecer con un buen grado de precision el valor que se desea buscar.

Este seria el polinio para aplicar a la funcion en grado n teniendo presente que h= x-a:

$f(x_{i+1})=f(x_i)+f'(x_i)h+\frac{f''(x_i)h^2}{2!}+\frac{f'''(x_i)h^3}{3!}+...+\frac{f^n(x_i)h^n}{n!}+R_n$

Donde Rn esta dado en la siguiente manera:

$R_n=\frac{f^{n+1}(\xi)h^{n+1}}{n+1!}$

```python
from sympy.codegen.fnodes import false
from sympy import lambdify,symbols,diff,exp,printing,latex,cos,plot,sin,Symbol
from sympy.plotting import plot
from IPython.display import display, Latex
from prettytable import PrettyTable
import matplotlib.pyplot as plt

import numpy as np

import math 
x = Symbol ('x')
a = Symbol ('a')
n = 5             # Cantidad de grados de f(x)
ao=(5*math.pi)/6  #Centrado en a 
xo =math.pi       # Evaluado en x


def f(x):         # Retorna cualquier funcion de F
  return cos(x)   # Funcion a Evaluar

def fact(n):      # Retorina cualquier factorial de n
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
#-----------------------------------------------------------------------------------------
# La siguiente funcion usa diff para derivar f(x) en funcion de x, en el grado n
# Sin embargo al retornar la funcion dfx(n-1)+.... vuelve a integrar hasta que n sea 0
# Por lo tanto se puede aplicar las veces que se requiera a la funcion F.
#-----------------------------------------------------------------------------------------

def dfx(n,Rn): # Toma el f(x) y genera la sumatoria de las derivadas n veces 
  if n==0 :
    return  f(a)
  else:
              # retorna la sumatoria de la serie Taylor ya simplificada
    return (((diff(f(a),a,n))*((x-a)**(n)))/(fact(n))+((diff(f(a),a,Rn))*((x-a)**(Rn)))/(fact(Rn)))+dfx(n-1,Rn)
    


def fxTaylor(ao,xo,rdfx):                # sumatoria de f(x)+ f^n divido en n! multiplicado por (x-x0)^n teniendo presente que se obtiene la funcion dfx(n) donde n es el grado 
    result = lambdify([a,x],rdfx)   # en la funcion rdfx tiene el Polinomio al cual se le aplican los valores de x y de a para obtener el resultado 
    return result(ao,xo)      # Retorno del resutado de la aproximacion 


tabla = PrettyTable(['Grado', 'Resultado','Error'])
listError=[]
labelG=[]

for i in range(6):
    #Calculo de los resutlados de la funcion conrespecto al grado 
    n=i
    rdfx=dfx(n,n+1)  
    rdgx=dfx(n,n+1)  
    err=dfx(n+1,n+2)
    error= fxTaylor(ao,xo,err)-fxTaylor(ao,xo,rdfx)
    print("\n","Evaluado en x: ",xo," Centrado en a:",ao," en el grado:",n," el resultado es: ", fxTaylor(ao,xo,rdfx))
    #Fin Calculo

    #Lista para la tabla 
    tabla.add_row([n, fxTaylor(ao,xo,rdfx),error])
    listError.append(error) 
    #Fin Lista 

    #Presentacion de Polinomio en formato tipo Libro
    formula = printing.latex(rdfx) 
    display(Latex(r"$$"+formula+"$$")) # para mostrar en pantalla en formato de formula 
    #Fin Formato
    
    # Zona Graficador de f(x) y de la funcion g(a) con el polinio encontrado 
    gy=[]
    x_g=np.linspace(ao ,xo,100)     #Intervalo de la linea de la grafica  
    try:
      g = rdgx.subs(a, ao)          #subtraccion de la variable 'a' asignado valor con try para saltar ecuaciones que no tengan la variable 
    except:
      g = rdgx


    for i in range(len(x_g)):       # Inicio Ciclo para Asignacion de lineas en la grafica 
      try:
       gy.append(g.subs(x, x_g[i])) #subtraccion de la variable 'x' asignado valor con try para saltar ecuaciones que no tengan la variable 
      except:
       gy.append(g)

    labelG.append("Grado "+str(n))  #Label de la grafica 
    plt.legend(labelG)              

    plt.plot(x_g,gy)                #Iteracion de la funcion de gy con respecto a x para graficar por ciclo
    plt.grid(True)

    if(len(labelG)==6):             #Agregar la funcion original a la grafica 
      plt.plot(xo,cos(xo))
      labelG.append("f(x)="+str(f(x)))
      plt.legend(labelG)
    #Fin Zona Graficador

  #Fin Ciclo de los 6 Grados 

# Impresion de la lista en formato tabla con el resultado y errores
print(tabla)

#Graficador de Error 
plt.figure()
plt.plot(range(0,n+1,1),listError)
plt.legend(["Error"])
plt.grid(True)




```
