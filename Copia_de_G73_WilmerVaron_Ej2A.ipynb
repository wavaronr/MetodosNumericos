#**Revisi√≥n te√≥rica de las Integraci√≥n num√©rica**

**La Regla del trapecio compuesta**

 Es un m√©todo num√©rico utilizado para aproximar la integral de una funci√≥n en un intervalo dado. La idea principal es dividir el intervalo en varios subintervalos y aproximar el √°rea bajo la curva en cada subintervalo como un trapecio. Luego, se suman todas las √°reas de los trapecios para obtener una aproximaci√≥n de la integral en el intervalo completo.

La f√≥rmula para la Regla del trapecio compuesta se ve as√≠:

$I\cong\frac{(b-a)}{2n}(f(x_0=a)+2\overset{n-1}{\underset{i=1}{\sum f(x_i)}}+f(x_n=b))$

Donde $a$ y $b$ son los l√≠mites inferior y superior del intervalo,$n$ es el n√∫mero de subintervalos y $‚Ñé$ es la longitud de cada subintervalo

**La Regla de Simpson 1/3 compuesta**

Es otro m√©todo num√©rico utilizado para aproximar la integral de una funci√≥n. Al igual que la Regla del trapecio compuesta, divide el intervalo en subintervalos, pero aproxima el √°rea bajo la curva en cada subintervalo utilizando una f√≥rmula m√°s precisa basada en una interpolaci√≥n cuadr√°tica.

La f√≥rmula para la Regla de Simpson 1/3 compuesta se expresa como:

$I\cong\frac h3\left[f(x_0)+4\overset{n-1}{\underset{i=1,3,5...}{\sum f(x_i)}}+2\overset{n-2}{\underset{j=2,4,6,..}{\sum f(x_i)}}+f(x_j)+f(x_n)\right]$

**La Regla de Simpson 3/8 compuesta**

Es un m√©todo num√©rico utilizado para aproximar la integral de una funci√≥n en un intervalo dado. Este m√©todo es una extensi√≥n de la Regla de Simpson 1/3 compuesta y proporciona una mayor precisi√≥n en la aproximaci√≥n.

La f√≥rmula de la Regla de Simpson 3/8 compuesta es la siguiente:

$I\cong\frac{3h}8\left[f(x_0)+3\overset n{\underset{i=0}{\sum f(x_{3i+1})}}+3\overset{\frac n3-1}{\underset{i=0}{\sum f(x_{3i+1})}}+2\overset{\frac n3-2}{\underset{i=0}{\sum f(x_{3i+1})}}+f(x_n)\right]$

En esta f√≥rmula, se divide el intervalo de integraci√≥n en subintervalos de igual tama√±o, donde 'n' es el n√∫mero de subintervalos y 'h' es la longitud de cada subintervalo. Se eval√∫a la funci√≥n en los extremos de cada subintervalo $(x_0, x_i, x_n)$ y se realiza la suma ponderada de estos valores utilizando los coeficientes adecuados $(3/8, 3/8, 2/8, 1/8)$para obtener una aproximaci√≥n de la integral total.


**La Cuadratura de Gauss** 
Es un m√©todo num√©rico utilizado para aproximar la integral de una funci√≥n en un intervalo determinado. Este m√©todo se basa en una f√≥rmula de integraci√≥n que utiliza puntos y pesos espec√≠ficos para obtener una aproximaci√≥n precisa.

En el caso de la Cuadratura de Gauss de 2 puntos, se seleccionan dos puntos dentro del intervalo de integraci√≥n, y se asignan pesos espec√≠ficos a cada punto. La f√≥rmula de la Cuadratura de Gauss de 2 puntos es la siguiente:

$I\cong\frac{(b-a)}2\left[w_1\ast f(\frac{(b+a)}2-\frac{(b-a)}2\sqrt{\frac13})+w_2\ast f(\frac{(b+a)}2+\frac{(b-a)}2\sqrt{\frac13})\right]$

En esta f√≥rmula, $a$ y $b$ representan los l√≠mites inferior y superior del intervalo de integraci√≥n respectivamente. Los puntos $x_1$ y $x_2$ est√°n dados por la f√≥rmula mencionada, y los pesos $w_1$ y $w_2$ son asignados de manera espec√≠fica para garantizar la precisi√≥n de la aproximaci√≥n.

Por otro lado, la Cuadratura de Gauss de 3 puntos sigue el mismo principio, pero en este caso se utilizan tres puntos y pesos correspondientes. La f√≥rmula de la Cuadratura de Gauss de 3 puntos es la siguiente:

$I\cong\frac{(b-a)}2\left[w_1\ast f(\frac{(b+a)}2-\frac{(b-a)}2\sqrt{\frac35})+w_2\ast f(\frac{(b+a)}2)+w_3\ast f(\frac{(b+a)}2+\frac{(b-a)}2\sqrt{\frac35})\right]$

```python
import numpy as np
import math
from tabulate import tabulate
from prettytable import PrettyTable

ùëõ_1 = 12 # intervalo asignado por el ejercici
n_2 = 15 # intervalo asignado por el ejercici
a=0
b= (np.pi/2)
I_Geogebra=0.9999999999987 #Valor calculado en Geogebra 
def f(x):         # Retorna cualquier funcion de F
  return  1/(1+np.sin(x))   # Funcion a Evaluar

# Verificador del error versus a constante calculada en Geogebra 
def errorIntegral(I_Geogebra,I):
  return ((abs(I_Geogebra-I))/I_Geogebra)*100


def trapecioCompuesto (a, b, n):
    h = (b - a) / n  # Ancho de cada subintervalo
    suma = f(a) + f(b)  # Suma de los valores en los extremos

    # Suma de los valores en los puntos interiores
    for i in range(1, n):
        suma += 2 * f(a + i * h)
    # Aplicar la f√≥rmula final
    I = (h / 2) * suma
    return I




def simpsonUnTercioCompuesta(a, b, n):

     # Definici√≥n de par√°metros
    
    h = (b - a) / n  # Tama√±o de cada subintervalo

    # Evaluaci√≥n de la funci√≥n en los extremos del intervalo
    fx0 = f(a)  # Valor de la funci√≥n en a
    fxn = f(b)  # Valor de la funci√≥n en b

    # Inicializaci√≥n de las sumatorias
    sumai = 0  # Sumatoria para los t√©rminos con √≠ndices impares
    sumap = 0  # Sumatoria para los t√©rminos con √≠ndices pares

    # C√°lculo de las sumatorias
    for i in range(1, n, 2):
        sumai += f(a + i * h)  # Sumatoria para los t√©rminos con √≠ndices impares
    for i in range(2, n, 2):
        sumap += f(a + i * h)  # Sumatoria para los t√©rminos con √≠ndices pares

    # C√°lculo de la aproximaci√≥n de la integral utilizando la f√≥rmula de Simpson 1/3 compuesta
    I = h / 3 * (f(a) + 4 * sumai + 2 * sumap + f(b))

    return I


def simpsonTresOctavosCompuesta(a, b, n):
    
    h = (b - a) / n  # Tama√±o de cada subintervalo

    # Evaluaci√≥n de la funci√≥n en los extremos del intervalo
    fx0 = fx=(a)  # Valor de la funci√≥n en a
    fxn = f(b)  # Valor de la funci√≥n en b

    # Inicializaci√≥n de las sumatorias
    suma1 = 0  # Primera sumatoria
    suma2 = 0  # Segunda sumatoria
    suma3 = 0  # Tercera sumatoria

    # C√°lculo de las sumatorias
    for i in range(1, n, 3):
        suma1 += f(a + i * h)  # Primera sumatoria
    for i in range(2, n, 3):
        suma2 += f(a + i * h)  # Segunda sumatoria
    for i in range(3, n, 3):
        suma3 += f(a + i * h)  # Tercera sumatoria

    # C√°lculo de la aproximaci√≥n de la integral utilizando la f√≥rmula de Simpson 3/8 compuesta
    I = 3 * h / 8 * (f(a) + 3 * suma1 + 3 * suma2 + 2 * suma3 + f(b))

    return I

def gaussDosPuntos(a, b):
    
    x1 = -0.5773502691896257  # coneficiente para x0
    x2 = 0.5773502691896257  #  coneficiente para x1

    # Transformaci√≥n del intervalo [a, b] al intervalo [-1, 1]
    t1 = ((b - a) * x1 + (b + a)) / 2
    t2 = ((b - a) * x2 + (b + a)) / 2

    # Aplicaci√≥n de la Cuadratura de Gauss de 2 puntos
    I = ((b - a) / 2) * (f(t1) + f(t2))
    return I

def gaussTresPuntos(a, b):
    #factores de argunmento y pesos de la formula de gauss
    x1 = -0.7745966692414834  # Ra√≠z de 3/5
    x2 = 0
    x3 = 0.7745966692414834  # Ra√≠z de 3/5
    c1=5/9 
    c2=8/9
    c3=5/9

    # Transformaci√≥n del intervalo [a, b] al intervalo [-1, 1]
    t1 = ((b - a) * x1 + (b + a)) / 2
    t2 = ((b - a) * x2 + (b + a)) / 2
    t3 = ((b - a) * x3 + (b + a)) / 2

    # Aplicaci√≥n de la Cuadratura de Gauss de 3 puntos
    I = ((b - a) / 2) * (c1 * f(t1) + c2 * f(t2) + c3 * f(t3))
    
    return I



#ini Cuadro con resultados por metodo numeros de intervalos y rango de error 
TPrimeraF = PrettyTable(["Metodo","n","Integral aprox","Error(%)"])
TPrimeraF.add_row(["Trapecio Compuesto ",n_1,trapecioCompuesto (a, b, n_1),errorIntegral(I_Geogebra,trapecioCompuesto (a, b, n_1))])
TPrimeraF.add_row(["Trapecio Compuesto ",n_2,trapecioCompuesto (a, b, n_2),errorIntegral(I_Geogebra,trapecioCompuesto (a, b, n_2))])
TPrimeraF.add_row(["Simpson 1/3 Compuesta ",n_1,simpsonUnTercioCompuesta (a, b, n_1),errorIntegral(I_Geogebra,simpsonUnTercioCompuesta (a, b, n_1))])
TPrimeraF.add_row(["Simpson 1/3 Compuesta ",n_2,simpsonUnTercioCompuesta (a, b, n_2),errorIntegral(I_Geogebra,simpsonUnTercioCompuesta (a, b, n_2))])
TPrimeraF.add_row(["Simpson 3/8 Compuesta ",n_1,simpsonTresOctavosCompuesta (a, b, n_1),errorIntegral(I_Geogebra,simpsonTresOctavosCompuesta (a, b, n_1))])
TPrimeraF.add_row(["Simpson 3/8 Compuesta ",n_2,simpsonTresOctavosCompuesta (a, b, n_2),errorIntegral(I_Geogebra,simpsonTresOctavosCompuesta (a, b, n_2))])
TPrimeraF.add_row(["Gauss Dos Puntos ","2 Puntos",gaussDosPuntos(a, b),errorIntegral(I_Geogebra,gaussDosPuntos(a, b))])
TPrimeraF.add_row(["Gauss Tres Puntos ","3 Puntos",gaussTresPuntos(a, b),errorIntegral(I_Geogebra,gaussTresPuntos(a, b))])


print(TPrimeraF)
# fin Cuadro 

# verificacion de minimo error para analisis
min_value = min(errorIntegral(I_Geogebra,trapecioCompuesto (a, b, n_1)),
                errorIntegral(I_Geogebra,trapecioCompuesto (a, b, n_2)),
                errorIntegral(I_Geogebra,simpsonUnTercioCompuesta (a, b, n_1)),
                errorIntegral(I_Geogebra,simpsonUnTercioCompuesta (a, b, n_2)),
                errorIntegral(I_Geogebra,simpsonTresOctavosCompuesta (a, b, n_1)),
                errorIntegral(I_Geogebra,simpsonTresOctavosCompuesta (a, b, n_2)),
                errorIntegral(I_Geogebra,gaussDosPuntos(a, b)),
                errorIntegral(I_Geogebra,gaussTresPuntos(a, b))
                )
#Organizacion minimo
sorted_values = sorted([
                errorIntegral(I_Geogebra,trapecioCompuesto (a, b, n_1)),
                errorIntegral(I_Geogebra,trapecioCompuesto (a, b, n_2)),
                errorIntegral(I_Geogebra,simpsonUnTercioCompuesta (a, b, n_1)),
                errorIntegral(I_Geogebra,simpsonUnTercioCompuesta (a, b, n_2)),
                errorIntegral(I_Geogebra,simpsonTresOctavosCompuesta (a, b, n_1)),
                errorIntegral(I_Geogebra,simpsonTresOctavosCompuesta (a, b, n_2)),
                errorIntegral(I_Geogebra,gaussDosPuntos(a, b)),
                errorIntegral(I_Geogebra,gaussTresPuntos(a, b))                
                ])
#El numero de veces que es mas grande del error  
diff = sorted_values[-1] / sorted_values[0]
 
print("El valor m√≠nimo es:", min_value)
print("La Cantidad de veces que es mas grande el error encontrado es:", diff)
```
