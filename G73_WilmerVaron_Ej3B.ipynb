# Metodo Bisección

El presente metodo corresponde a la busque de la raiz $f(x)=0$ cuando esta no es evidente en los metodos convencionales, se procede a aplicar el presente, por lo tanto se toma de manera arbitraria dos puntos uno a y otro b con signo contrario con el fin de aplicar el siguiente concepto

$$c=\frac{a+b}2\\ $$

Donde el **c** viene a ser nuestro valor raiz, como podremos daber que hemos encontrado el valor correcto mas cercano a la raiz, la forma de verificacion es aplicando el siguiente concepto $f(c)=0$ si el valor al aplicar el punto en la funcion es cero hemos encontrado nuestro valor, pero si solo es un numero diferente operamos de la siguiente manera $$f(c)*f(a)<0$$

Entonces asumimos que el c es el valor menor y ahora le asignamos el valor de "c" a "a" y nuevamente buscamos el punto medio de ese nuevo intervalo ahora en el caso de que al aplicar sea $$f(c)*f(a)>0$$

Entonces le atribuimos el valor a la variable "b" dado a que en el evento que esta multiplicacion sea cero hemos encontrado la raiz o en un dado caso que estemos dispuestos a tolerar un punto de error con cierto nivel de aproximacion a la raiz y el error esta dado por la diferencia entre el punto negativo b y a en su valor absoluto

```python
from IPython.display import display, Latex
from sympy import printing,Symbol,diff
import numpy as np
import pandas as pd
a=2 #Punto del intervalo positivo
b=3 #Punto del intervalo negativo
i=0 # Cantidad de acercacmiento al para llegar al púnto cero
tol=10**(-6) # Valor de tolerancia de error 
error=1
errorBise=[] #Lista de almacenamiento de los errores 
AproxBise=[] #Valor de la aproximacion a cero 
x= Symbol('x') # x simbolica para imprimir funcion por display 


def f(x):         # Retorna cualquier funcion de F
  return (x**3)-3.4*(x**2)+2.5*x-0.7

def displayf(f):
  formula = printing.latex(f) 
  return display(Latex(r"$$"+formula+"$$")) # para mostrar en pantalla en formato de libro 

while (i<100 and error >tol):
  c=(a+b)/2           # Valor c buscado como punto medio
  AproxBise.append(c) # agrega el ultimo valor de c
  if f(a)*f(c)<0:     #Al realizar la multiplicacion busca que el valor negativo buscando el valor cada vez mas pequeño
    b=c
  else:
    a=c               #Si a es menor lo combierte en el nuevo c 
  i+=1
  error=b-a           #Verificacion del error 
  errorBise.append(error) # almacenamiento el historico de erres 

displayf(f(x)) # Mostrar la formula den formato de libro 
print(f'La aproximacion por M Biseccion de la raiz es:{c} con error {error} en {i}')

dicBise={'Aprox. Bisección':AproxBise, 'Error':errorBise}
dicBiseT=pd.DataFrame.from_dict(dicBise)
print(dicBiseT)

dicBiseT['Aprox. Bisección'].plot()

```

#Metodo  Newton–Raphson

El presente metodo al igual que el anterior busca establecer la raiz de $f(x)=0$ por medio de la busqueda de la un punto c el cual es dado por la tangente entre la curva y el eje x para el cual se presenta el acercamiento aplicando el siguiente concepto 

$$ x_{i+1}=x_i-\frac{f(x_i)}{f'(x_i)}\\ $$

donde el valor nuevo a cada iteracion mas cercano al valor de la raiz, es impoortante aclarar que $ x_0 $ es un valor arbitrario tomado en la grafica como punto de partida a la iteracion del metodo y el establecimiento del error se calcula hallando el valor absoluto de la diferencia entre $x_{i+1}-x_i$

Adicional para poder aplicar el método se debe tener presente la convergencia, precisando las condiciones que la $f(x)$ tiene su $f''(x)$ y es continua además esta tiene diferente signo en ambos puntos del intervalo y siempre y cuando la $f'(x)$ y $f''(x)$ sean diferentes de cero en el intervalo $[a,b]$

```python
from IPython.display import display, Latex
from sympy import printing,Symbol,diff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
i=0
tol=10**(-6) # Valor de tolerancia de error 
error=1
errorNR=[]
aproxNR=[3]
x= Symbol('x')
def f(x):         # Retorna cualquier funcion de F
  return (x**3)-3.4*(x**2)+2.5*x-0.7

def df(x):
  return diff(f(x),x,1) #Derivada de la funcion f(x)

while (i<100 and error >tol):
  xi=aproxNR[-1] # asignacion del ultimo valor de lista 
  fx=f(xi) 
  dfx=df(x)
  dfx=dfx.subs(x,xi) # asignacion de valor a x la derivada de la funcion 
  X=xi-(fx/dfx)      # Aplicacion de la divicion de la funcion en su deriva menos el valor de x 
  
  aproxNR.append(X) # almacenamiento del x mas pequeño un una lista
  i+=1
  error=abs(X-xi) # Establecimiento del error del ultimo x hallado 
  errorNR.append(error) # almacenamiento en una lista el historial de errores 
print(f'La aproximacion por M Biseccion de la raiz es:{X} con error {error} en {i}') 


dicNR={'Aprox. NR':aproxNR[1:], 'Error':errorNR } 
dfNR=pd.DataFrame.from_dict(dicNR)
print(dfNR) #Impresion de la tabla consutruida con el diccionario alimentado con las listas almacenadas
gy=dfNR['Aprox. NR']
x_g=np.linspace(1 ,len(gy),len(gy))  

plt.plot(x_g,gy)  # grafica de el valor de tendencia del error 


```

```python
#Codigo para establecer conslusiones de los metodos utilizados 
MetodoNewtonRaphson=2.51735525472014  # consutante establecida como valor de x para encontrar la raiz con MetodoNewtonRaphson
MetodoBiseccion=2.5173559188842773    # consutante establecida como valor de x para encontrar la raiz con MetodoBiseccion

def f(x):         # Retorna cualquier funcion de F
  return ((x**3)-3.4*(x**2)+2.5*x-0.7)

if(f(MetodoNewtonRaphson)<f(MetodoBiseccion)): # Comparacion de los metodos al pasar los valores encontrados por su repectiva funcion 
  print("El metodo mas cercano a cero es Metodo Newton Raphson con ",f(MetodoNewtonRaphson), " y Metodo Biseccion con",f(MetodoBiseccion))
  print("El valor del Metodo Newton Raphson",format(f(MetodoBiseccion)/f(MetodoNewtonRaphson), ','), "mas pequeño que Metodo Biseccion")
else:
  print("El metodo mas cercano a cero es Metodo Biseccion con ",f(MetodoBiseccion), " y Metodo Newton Raphson con",f(MetodoNewtonRaphson))
  print("El valor del Metodo Biseccion",format(f(MetodoNewtonRaphson)/f(MetodoBiseccion),','), "mas pequeño que Metodo Newton Raphson")


```
