#**Revisión teórica de las Diferenciación numérica**

Para el presente tema debemos recordar, temas anteriores como las teorías de errores mas precisamente hablando del teorema de Taylor el cual por medio de sumatoria polinomios que acercamos a la función permitiendo encontrar el valor del función, es de tener en cuenta que estos polinomios contienen las derivadas de la función en distintos grados, por lo tanto si meditamos en la misma podemos encontrar una relación veamos la formula del teorema.


$f(x_{i+1})=f(x_i)+f'(x_i)h+\frac{f''(x_i)h^2}{2!}+\frac{f'''(x_i)h^3}{3!}+...+\frac{f^n(x_i)h^n}{n!}+R_n$

Entonces podemos notar que si tomamos la relacion que tiene el polinomio en su primero o segundo grado en ambos caso podemos hallar $f'(x)$, haciendo el despeje de la ecuacion para encontrar en lugar de $f(x_{i+1})$ buscamos la derivada de la funcion  

Cuando vamos a encontrar la funcion progresiva nos quedaria de la siguiente forma:

$f'(x_0)=\frac{f(x_0+h)-f(x_0)}h$

Ahora si se toma las diferencias finitas centradas 

$f'(x_0)=\frac{f(x_0+h)-f(x_0-h)}{2h}$

Por ultimo si se toma las diferencias finitas regresivas

$f'(x_0)=\frac{f(x_0)-f(x_0-h)}h$

Bueno de igual forma podemos notar que tambien podemos encontrar $f''(x)$ por lo tanto si aplicamos el mismo principio de la primera derivada logramos establecer la formula que nos permite acercarnos al valor de la derivada segunda bien sea progresiva, centrada o regresiva.

Para la progresiva quedaria de la siguiente manera 

$f''(x_0)=\frac{f(x_0)-2f(x_0+h)+f(x_0+2h)}{h^2}$

Para el caso del metodo central 

$f''(x_0)=\frac{f(x_0-h)-2f(x_0)+f(x_0+h)}{h^2}$

Po ultimo en el caso del regresivo 

$f''(x_0)=\frac{f(x_0-2h)-2f(x_0-h)+f(x_0)}{h^2}$

```python
from numpy import exp
import math
from tabulate import tabulate
from prettytable import PrettyTable

# Datos suministrados por el ejericio 
x=2.2
h1=0.5
h2=0.05
h3=0.01

def f(x): #funcion del ejecicio
    return  math.cosh(x)

def fPrimeraForward(x,h): # Aplicacionprimera derivada hacia adelante
    return (f(x+h)-f(x))/h

def fPrimeraCentral(x,h): # Aplicacionprimera derivada hacia Central
  return (f(x+h)-f(x-h))/(2*h)

def fPrimeraBackward(x,h):  # Aplicacionprimera derivada hacia Atras
  return (f(x)-f(x-h))/h

DR= math.sinh(x)
#evaluacion en las diferentes h propuestas por el ejercicio
fx1=fPrimeraForward(x,h1)
fx2=fPrimeraForward(x,h2)
fx3=fPrimeraForward(x,h3)

fx4=fPrimeraCentral(x,h1)
fx5=fPrimeraCentral(x,h2)
fx6=fPrimeraCentral(x,h3)

fx7=fPrimeraBackward(x,h1)
fx8=fPrimeraBackward(x,h2)
fx9=fPrimeraBackward(x,h3)


def errorDf(DR,f): #Error relativo porcentual 
  return (abs(DR-f)/DR)*100


#Tabla de con la evaluacion de la derivada en las h y con los errores calculados
TPrimeraF = PrettyTable(["Esquema","Tamaño h","Derivado aprox","Error(%)"])
TPrimeraF.add_row(["Adelante",h1,fPrimeraForward(x,h1),errorDf(DR,fx1)])
TPrimeraF.add_row(["Adelante",h2,fPrimeraForward(x,h2),errorDf(DR,fx2)])
TPrimeraF.add_row(["Adelante",h3,fPrimeraForward(x,h3),errorDf(DR,fx3)])
TPrimeraF.add_row(["Central",h1,fPrimeraCentral(x,h1),errorDf(DR,fx4)])
TPrimeraF.add_row(["Central",h2,fPrimeraCentral(x,h2),errorDf(DR,fx5)])
TPrimeraF.add_row(["Central",h3,fPrimeraCentral(x,h3),errorDf(DR,fx6)])
TPrimeraF.add_row(["Atras",h1,fPrimeraBackward(x,h1),errorDf(DR,fx7)])
TPrimeraF.add_row(["Atras",h2,fPrimeraBackward(x,h2),errorDf(DR,fx8)])
TPrimeraF.add_row(["Atras",h3,fPrimeraBackward(x,h3),errorDf(DR,fx9)])

print(TPrimeraF)
# verificacion de minimo error para analisis
min_value = min(errorDf(DR,fx1),
                errorDf(DR,fx2),
                errorDf(DR,fx3),
                errorDf(DR,fx4),
                errorDf(DR,fx5),
                errorDf(DR,fx6),
                errorDf(DR,fx7),
                errorDf(DR,fx8),
                errorDf(DR,fx9),
                )
#Organizacion minimo
sorted_values = sorted([
                errorDf(DR,fx1),
                errorDf(DR,fx2),
                errorDf(DR,fx3),
                errorDf(DR,fx4),
                errorDf(DR,fx5),
                errorDf(DR,fx6),
                errorDf(DR,fx7),
                errorDf(DR,fx8),
                errorDf(DR,fx9),
                ])

#El numero de veces que es mas grande del error  
diff = sorted_values[-1] / sorted_values[0]

print("El valor mínimo es:", min_value)
print("La Cantidad de veces que es mas grande el error encontrado es:", diff)
```
